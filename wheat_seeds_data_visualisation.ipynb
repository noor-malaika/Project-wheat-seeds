```python
# ANDREW CURVES

import pandas as pd
import matplotlib.pyplot as plt

from pandas.plotting import andrews_curves

# Look pretty...
plt.style.use('ggplot')


#
# Load up the Seeds Dataset into a Dataframe
# 
df = pd.read_csv("normalised_with_col_names.csv", index_col='id')
print(df.columns)



#
#  get rid of the 'area' and 'perimeter' features
# 
df2 = df.drop(['area','perimeter'], axis=1)
print(df.head())

#
# Plot a parallel coordinates chart grouped by
# the 'wheat_type' feature. Be sure to set the optional
# display parameter alpha to 0.4
# 
plt.figure()
andrews_curves(df2, 'wheat_type', alpha=0.4)

plt.figure()
andrews_curves(df, 'wheat_type', alpha=0.4)

plt.show()
```

```python
# CORRELATION

import pandas as pd
import matplotlib.pyplot as plt


#
# Load up the Seeds Dataset into a Dataframe
# 
df = pd.read_csv("normalised_with_col_names.csv", index_col='id')


#
# Compute the correlation matrix of the dataframe
# 
c = df.corr()
print(c)

#
# Graph the correlation matrix using imshow
# 
plt.imshow(c, cmap=plt.cm.Blues, interpolation='nearest')

plt.colorbar()
tick_marks = [i for i in range(len(df.columns))]
plt.xticks(tick_marks, df.columns, rotation='vertical')
plt.yticks(tick_marks, df.columns)

plt.show()
```

```python
# HISTOGRAMS

import pandas as pd
import matplotlib.pyplot as plt

# Look pretty...
plt.style.use('ggplot')


#
# Load up the Seeds Dataset into a Dataframe
# 
df = pd.read_csv("normalised_with_col_names.csv")
print(df.columns)

#
# Create a slice of the dataframe
# that only includes the 'area' and 'perimeter' features
# 
s1 = df[['area','perimeter']]
print(s1.head())

#
# Create another slice of the dataframe 
# that only includes the 'groove' and 'asymmetry' features
# 
s2 = df[['groove','asymmetry']]
print(s2.head())


#
# Create a histogram plot using the first slice,
# and another histogram plot using the second slice.
# 
s1.plot.hist(alpha=0.75)
s2.plot.hist(alpha=0.75)

# Display the graphs:
plt.show()
    # which feature has more variance?
pd.DataFrame.var(s2.groove)
pd.DataFrame.var(s2.asymmetry)
```

```python
# PARALLEL COORDINATES

import pandas as pd
import matplotlib.pyplot as plt

from pandas.plotting import parallel_coordinates

# Look pretty...
plt.style.use('ggplot')


#
# Load up the Seeds Dataset into a Dataframe
# 
df = pd.read_csv("normalised_with_col_names.csv", index_col='id')
print(df.columns)



#
#  get rid of the 'area' and 'perimeter' features
# 
df.drop(['area','perimeter'], axis=1, inplace=True)
print(df.head())

#
# Plot a parallel coordinates chart grouped by
# the 'wheat_type' feature. 
# 
plt.figure()
parallel_coordinates(df, 'wheat_type', alpha=0.4)


plt.show()
```

```python
# SCATTER PLOT(2d)

import pandas as pd
import matplotlib.pyplot as plt

# Look pretty...
plt.style.use('ggplot')


#
# Load up the Seeds Dataset into a Dataframe
# 
df = pd.read_csv("normalised_with_col_names.csv")
print(df.columns)


#
# Create a 2d scatter plot that graphs the
# area and perimeter features
# 
a = ['kama', 'rosa', 'canadian']
is_x, is_y, ive_x, ive_y, ivi_x, ivi_y = [], [], [], [], [], []

for e,i in enumerate(df["wheat_type"]):
    if(i == a[0]):
        is_x.append(df["area"][e])
        is_y.append(df["perimeter"][e])
    elif(i == a[1]):
        ive_x.append(df["area"][e])
        ive_y.append(df["perimeter"][e])
    elif(i == a[2]):
        ivi_x.append(df["area"][e])
        ivi_y.append(df["perimeter"][e])

plt.plot(is_x, is_y, 'b.', label = "Kama")
plt.plot(ive_x, ive_y, 'g.', label = "Rosa")
plt.plot(ivi_x, ivi_y, 'r.', label = "Canadian")

plt.legend()
plt.show()

#
# Create a 2d scatter plot that graphs the
# groove and asymmetry features
# 

for e,i in enumerate(df["wheat_type"]):
    if(i == a[0]):
        is_x.append(df["groove"][e])
        is_y.append(df["asymmetry"][e])
    elif(i == a[1]):
        ive_x.append(df["groove"][e])
        ive_y.append(df["asymmetry"][e])
    elif(i == a[2]):
        ivi_x.append(df["groove"][e])
        ivi_y.append(df["asymmetry"][e])

plt.plot(is_x, is_y, 'b.', label = "Kama")
plt.plot(ive_x, ive_y, 'g.', label = "Rosa")
plt.plot(ivi_x, ivi_y, 'r.', label = "Canadian")

plt.legend()

plt.show()

#
# Create a 2d scatter plot that graphs the
# compactness and width features
# 

for e,i in enumerate(df["wheat_type"]):
    if(i == a[0]):
        is_x.append(df["compactness"][e])
        is_y.append(df["width"][e])
    elif(i == a[1]):
        ive_x.append(df["compactness"][e])
        ive_y.append(df["width"][e])
    elif(i == a[2]):
        ivi_x.append(df["compactness"][e])
        ivi_y.append(df["width"][e])

plt.plot(is_x, is_y, 'b.', label = "Kama")
plt.plot(ive_x, ive_y, 'g.', label = "Rosa")
plt.plot(ivi_x, ivi_y, 'r.', label = "Canadian")

plt.legend()


plt.show()
```

```python
# SCATTER PLOT(3d)

import pandas as pd
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D

# Look pretty...
plt.style.use('ggplot')


#
# Load up the Seeds Dataset into a Dataframe
# 
df = pd.read_csv("normalised_with_col_names.csv")



fig = plt.figure()

#
# Create a new 3D scatter plot using the area,
# perimeter and asymmetry features. 
# 
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Area')
ax.set_ylabel('Perimeter')
ax.set_zlabel('Asymmetry')

ax.scatter(df.area, df.perimeter, df.asymmetry, c='red')

fig = plt.figure()
#
# Create a new 3D scatter plot using the width,
# groove and length features. 
# 
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('width')
ax.set_ylabel('groove')
ax.set_zlabel('length')

ax.scatter(df.width, df.groove, df.length, c='green')

plt.show()
```
